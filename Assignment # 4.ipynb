# Print Sum of N numbers

```python
# 1. We first take a value from the user using input().
# 2. Then, we initialize sum to 0.
# 3. Then, we loop from 0 to that number and add all the numbers.
# 4. Finally, we print the sum.

inp_value = int(input("Enter any number : " ))
sum = 0
for i in range(0,inp_value+1):
    sum += i
print(sum)
```

# Practice of Dictonary

```python
# 1. The user is asked to enter the name of the exam.
# 2. The name of the exam is stored in a variable called name_of_exam.
# 3. If the name of the exam is present in the dictionary, the corresponding value is printed.
# 4. If the name of the exam is not present in the dictionary, the message “Enter correct exam name” is printed.

name_of_exam = input("Enter exam name you want score of : ")

student_123 = {
    "Name" : "Mohsin",
    "Class": "A-I",
    "Percentage" : "60%",
    "Marks" : {
        "Quiz_1": 40,
        "Quiz_2" : 45,
        "Mid" : 50,
        "Final" : 55,
    }
}

if name_of_exam in student_123["Marks"]:
    print(student_123["Marks"][name_of_exam])

else:
    print("Enter correct exam name")
```

# checking for  Discount offer

```python
# Discount apply according to this given  below,
# brother-in-law – 30%
# loyalty – 15%
# volume – 10%
# standard – 5%

customer_29876 = {
     "first name": "David",
     "last name": "Elliott",
     "address": "4803 Wellesley St.",
     "discounts": ["standard", "volume", "loyalty"],
 }

# checking a condtion how many of the disccount  offer should be apply on this user
# 1. First, we check if the customer has the brother-in-law discount. If so, we add 0.30 to the discount.
# 2. Next, we check if the customer has the loyalty discount. If so, we add 0.15 to the discount.
# 3. Next, we check if the customer has the volume discount. If so, we add 0.10 to the discount.
# 4. Next, we check if the customer has the standard discount. If so, we add 0.05 to the discount.
# 5. Finally, we print the discount that was applied to the customer.

discount = 0

if "brother-in-law" in customer_29876["discounts"]:
        discount += 0.30
if "loyalty" in customer_29876["discounts"]:
        discount += 0.15
if "volume" in customer_29876["discounts"]:
        discount += 0.10
if "standard" in customer_29876["discounts"]:
        discount += 0.05
        
print(customer_29876["discounts"])
print("Discount applied to the customer_29876 is : ",  discount )
```
